Summary of Report

---


  The purpose of this project is to analyze the housing prices of homes posted on Zillow. The dataset used in this project is Zillow economics data, which contains information on various features of real estate and their corresponding prices. The aim of this project is to gain insights into the factors that influence housing prices.

  This project utilizes a dataset making real estate predictions from Zillow. The attributes include features such as the number of rooms, square feet, and age of home. The dataset is publicly available and is accessed through Kaggle.

  The primary goal of this project is to determine which housing features have the most significant impact on the price of a house. The questions that are addressed in this project include:

1. Which property features have the strongest correlation with housing prices?

The property feature with the strongest correlation to housing prices is square feet. The current analysis shows an R-squared value of 0.635, indicating that approximately 63.5% of the variation in housing prices can be explained by the variation in square feet.

2. How does the size of a house affect its price?

The current data reveals a significant correlation between the size of a house and its price. The regression coefficient of 332.278 suggests that for every increase in square foot, the price increases by $332.28. The summary statistics support this finding, with an R-squared value of 0.635, indicating that approximately 63.5% of the variability in housing prices can be explained by the variation in square feet.

3. How does the price of a home change over time?

The analysis of housing prices over time shows that for every additional year in the age of the home, the price decreases by \$2,397 in 2016 and $2,331.52 in 2017. The R-squared value suggests that approximately 42.1% of the variation in housing price is explained by the variation in the year they were built. The p-values are very close to 0, indicating a statistically significant relationship between the age of homes and the price of homes.

4. What is the relationship between the number of bedrooms and housing prices?

The regression coefficient indicates that for every additional bedroom, the price of the home increases by approximately $128,100 on average. The p-value is also statistically significant, suggesting a moderate correlation between the number of bedrooms and the price of a home.

5. What is the relationship between the number of bathrooms and housing prices?

The R-squared value of 0.571 indicates that approximately 57.1% of the variation in housing prices can be explained by the number of bathrooms a home has. This indicates a strong correlation between the number of bathrooms and the price of a home. The p-value suggests a statistically significant relationship between these two variables, indicating that the number of bathrooms is a determining factor for the price of a home.
